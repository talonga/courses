```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
import warnings

%matplotlib inline
warnings.filterwarnings('ignore')
```

```python
sns.lmplot(data=stats, x='InternetUsers', y='BirthRate', hue='IncomeGroup', fit_reg=False, size=8,
          scatter_kws={'s':100})
```
